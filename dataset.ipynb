```python
df=pd.DataFrame()
for img in UCMerced("data"):
    stat=ImageStat.Stat(img)
    df=df.append({
        'width':img.width,
        'height':img.height,
        'mean_r':stat.mean[0],
        'mean_g':stat.mean[1],
        'mean_b':stat.mean[2],
        'std_r':stat.stddev[0],
        'std_g':stat.stddev[1],
        'std_b':stat.stddev[2],
        'min_r':stat.extrema[0][0],
        'min_g':stat.extrema[1][0],
        'min_b':stat.extrema[2][0],
        'max_r':stat.extrema[0][1],
        'max_g':stat.extrema[1][1],
        'max_b':stat.extrema[2][1],
        }, ignore_index=True)
```
